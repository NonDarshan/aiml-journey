```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Numpy Array Basics

arr = np.array([10, 20, 30, 40, 50])
arr, arr.shape
```

```python
# Mean, Sum, Max, Min

arr_mean = arr.mean()
arr_sum = arr.sum()
arr_max = arr.max()
arr_min = arr.min()

arr_mean, arr_sum, arr_max, arr_min
```

```python
# Reshape a matrix

matrix = np.arange(9).reshape(3, 3)
matrix
```

```python
arr_plus_5 = arr + 5
arr_times_2 = arr * 2
arr_plus_5, arr_times_2
```

```python
data = {
    "Name": ["Aarav", "Reva", "Darshan", "Mihir"],
    "Math": [90, 75, 80, 95],
    "Science": [88, 72, 78, 89],
    "English": [82, 70, 85, 90]
}
df = pd.DataFrame(data)
df
```

```python
df.describe()
df["Math"]
df[df["Math"] >80 ]
```

```python
df["Total"] = df[["Math", "Science", "English"]].sum(axis=1)
df["Average"] = df["Total"]/3
#df

topper = df.loc[df["Total"].idxmax()]
topper
```

```python
plt.figure(figsize=(8, 4))
plt.bar(df["Name"], df["Total"])
plt.title("Total Scores")
plt.show()
```
